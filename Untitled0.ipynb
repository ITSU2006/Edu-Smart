```python
df = pd.read_excel('customer_orders_rawdata.xlsx', header=1)
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
print("Columns:\n", df.columns)
print("\nMissing Values:\n", df.isnull().sum())
print("\nDuplicates:", df.duplicated().sum())
```

```python
# Fill numeric missing values with median
df['quantity'] = df['quantity'].fillna(df['quantity'].median())

# Fill categorical missing values with mode
df['country'] = df['country'].fillna(df['country'].mode()[0])

# Drop rows where order_date is missing
df = df.dropna(subset=['order_date'])
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
df['signup_date'] = pd.to_datetime(df['signup_date'], errors='coerce')
```

```python
# Standardize product categories
df['product_category'] = df['product_category'].str.lower()
df['product_category'] = df['product_category'].replace({'clthng': 'clothes', 'elec.': 'electronics'})

# Standardize country names
df['country'] = df['country'].replace({'AUS': 'Australia', 'AU': 'Australia', 'Down Under': 'Australia'})

# Standardize returned/delivery status
df['returned'] = df['returned'].replace({1: 'Yes', 0: 'No'})
```

```python
df['product_price'] = pd.to_numeric(df['product_price'], errors='coerce')
df = df.dropna(subset=['product_price'])
```

```python
df['annual_income'] = df['annual_income'].astype(str)
df['annual_income'] = df['annual_income'].str.replace('AUD', '').str.replace(',', '')
df['annual_income'] = pd.to_numeric(df['annual_income'], errors='coerce')
```

```python
Q1 = df['product_price'].quantile(0.25)
Q3 = df['product_price'].quantile(0.75)
IQR = Q3 - Q1

df = df[(df['product_price'] >= Q1 - 1.5 * IQR) &
        (df['product_price'] <= Q3 + 1.5 * IQR)]
```

```python
df['total_amount'] = df['product_price'] * df['quantity']
```

```python
print("\nMissing Values After Cleaning:\n", df.isnull().sum())
print("\nFinal Shape:", df.shape)
```

```python
df.to_csv('cleaned_orders.csv', index=False)
print("\n✅ Cleaned dataset saved as cleaned_orders.csv")
```
